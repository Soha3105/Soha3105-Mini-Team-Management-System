```python
team_members = [{
        "name": "",
        "role": "Developer",
        "task": "Create login page",
        "status": "Pending"
    },
    {
        "name": "Priya",
        "role": "UI/UX Designer",
        "task": "Design dashboard layout",
        "status": "In Progress"
    },
    {
        "name": "Rahul",
        "role": "Backend Developer",
        "task": "Build REST API",
        "status": "Completed"
    },
    {
        "name": "Sneha",
        "role": "QA Engineer",
        "task": "Test user registration module",
        "status": "Pending"
    },
    {
        "name": "Karan",
        "role": "DevOps Engineer",
        "task": "Setup CI/CD pipeline",
        "status": "In Progress"
    },
    {
        "name": "Neha",
        "role": "Data Analyst",
        "task": "Analyze user engagement data",
        "status": "Completed"
    },
    {
        "name": "Sujal",
        "role": "Mobile App Developer",
        "task": "Develop Android login flow",
        "status": "Pending"
    },
    {
        "name": "Reena",
        "role": "Project Manager",
        "task": "Prepare project timeline",
        "status": "Completed"
    },
    {
        "name": "Rohan",
        "role": "Security Engineer",
        "task": "Perform vulnerability assessment",
        "status": "In Progress"
    },
    {
        "name": "Ananya",
        "role": "Content Writer",
        "task": "Write user documentation",
        "status": "Pending"
    }]

def add_member():
    name = input("Enter member name: ").strip()

    for member in team_members:
        if member["name"].lower() == name.lower():
            print("‚ùå Member already exists")
            return

    role = input("Enter role: ").strip()
    task = input("Enter task: ").strip()

    team_members.append({
        "name": name,
        "role": role,
        "task": task,
        "status": "Pending"
    })

    print("‚úÖ Team member added successfully")


def view_members():
    if not team_members:
        print("No team members found.")
        return

    for member in team_members:
        print("\nName:", member["name"])
        print("Role:", member["role"])
        print("Task:", member["task"])
        print("Status:", member["status"])

def update_status():
    name = input("Enter member name to update status: ").strip()

    for member in team_members:
        if member["name"].lower() == name.lower():
            print("\n1. Pending")
            print("2. In Progress")
            print("3. Completed")

            choice = input("Select status: ").strip()

            if choice == "1":
                member["status"] = "Pending"
            elif choice == "2":
                member["status"] = "In Progress"
            elif choice == "3":
                member["status"] = "Completed"
            else:
                print("‚ùå Invalid choice")
                return

            print("‚úÖ Status updated successfully")
            return

    print("‚ùå Member not found")

def progress_report():
    total = len(team_members)
    completed = sum(1 for m in team_members if m["status"] == "Completed")
    pending = total - completed
    percentage = (completed / total * 100) if total > 0 else 0

    print("\nüìä Team Progress Report")
    print("Total Team Members:", total)
    print("Completed Tasks:", completed)
    print("Pending Tasks:", pending)
    print(f"Completion Percentage: {percentage:.2f}%")

while True:
    print("\nTeam Management System")
    print("1. Add Team Member")
    print("2. View Team Members")
    print("3. Update Task Status")
    print("4. Team Progress Report")
    print("5. Exit")

    choice = input("Enter your choice: ").strip()

    if choice == "1":
        add_member()
    elif choice == "2":
        view_members()
    elif choice == "3":
        update_status()
    elif choice == "4":
        progress_report()
    elif choice == "5":
        print("Exiting program...")
        break
    else:
        print("‚ùå Invalid choice. Try again.")
def add_member():
    name = input("Enter member name: ").strip()

    for member in team_members:
        if member["name"].lower() == name.lower():
            print("‚ùå Member already exists")
            return

    role = input("Enter role: ").strip()
    task = input("Enter task: ").strip()

    team_members.append({
        "name": name,
        "role": role,
        "task": task,
        "status": "Pending"
    })


```
